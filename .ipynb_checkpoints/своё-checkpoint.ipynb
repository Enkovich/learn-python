```python
import requests

responce = requests.get('https://jsonplaceholder.typicode.com/posts')
responce = responce.json()

print(responce[0].keys(), '\n')

for i in range(len(responce)):
    print(f"{responce[i]['id']}. {responce[i]['title']}\n{responce[i]['body']}\n\n")
```
