<h3>Задание 3. Кратные числа</h3>

Из множества целых чисел от 1 до $N$ выделить множество $N_2$ чисел кратных 2, множество $N_3$ – кратных 3, множество $N_6$ – кратных 6 (т.е. кратные и 2 и 3), множество $N_{23}$ – кратных либо 2, либо 3.

````python
N = 100
N2 = set()
N3 = set()
N6 = set()
N23 = set()
for i in range(N):
    if i%2==0:
        N2.add(i)
    if i%3==0:
        N3.add(i)
    
N6 = N2.intersection(N3)
N23 = N2.symmetric_difference(N3)
print(N2, '\n')
print(N3, '\n')
print(N6, '\n')
print(N23, '\n')
````

<h3>Задание 4. Вхождения списка</h3>

Напишите программу на Python, чтобы найти список целых чисел, в котором ровно два раза встречаются числа девятнадцать и как минимум три раза встречаются числа пять. Вернуть True, иначе False

```
[19, 19, 15, 5, 3, 5, 5, 2] => True
[19, 15, 15, 5, 3, 3, 5, 2] => False
[19, 19, 5, 5, 5, 5, 5] => True
```

````python
def spisok(arr):
    s = dict(Counter(arr))
    if s[19]==2 and s[5]>=3:
        return True
    else:
        return False

print(spisok([19, 19, 15, 5, 3, 5, 5, 2]))
print(spisok([19, 15, 15, 5, 3, 3, 5, 2]))
print(spisok([19, 19, 5, 5, 5, 5, 5]))
````

<h3>Задание 5. Преобразование</h3>

Имеется коллекция отрицательных целых чисел. Напишите функцию, которая находит и возводит в квадрат все повторяющиеся элементы коллекции. Функция возвращает преобразованную таким образом коллекцию. Если повторяющихся элементов нет, функция возвращает False.

```
[-7, -3, -7, -4, -4, -7, -1, -1] => [-7, -3, 49, -4, 16, 49, -1, 1]
[-1, -2, -3, -4, -5] => False
```

````python
def preobrazovanie(arr):
    nums = list()
    new_arr = list()
    for i in range(len(arr)):
        if arr[i] not in nums:
            new_arr.append(arr[i])
            nums.append(arr[i])
        else:
            new_arr.append(arr[i]**2)
    if new_arr == arr:
        return False
    else:
        return new_arr
    
print(preobrazovanie([-7, -3, -7, -4, -4, -7, -1, -1]))
print(preobrazovanie([-1, -2, -3, -4, -5]))
````

<h3>Задание 6. Повторяющийся элемент</h3>

Напишите функцию, возвращающую первый повторяющийся элемент коллекции целых чисел. Верните -1, если таких элементов нет.

```
[1, 2, 3, 4, 4, 5] => 4
[1, 2, 3, 4] => -1
[1, 1, 2, 3, 3, 2, 2] => 1
```

````python
def povtor(arr):
    s = dict(Counter(arr))
    for el in s.keys():
        if s[el]>1:
            return el
    return -1

print(povtor([1, 2, 3, 4, 4, 5]))
print(povtor([1, 2, 3, 4]))
print(povtor([1, 1, 2, 3, 3, 2, 2]))
````
