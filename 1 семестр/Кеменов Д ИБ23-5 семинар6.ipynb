**Задание 1**

Дана строка, состоящая из слов, разделенных пробелами. Выведите все слова, встречающиеся в этой строке, на экран. Рядом выведите, сколько раз встречается слово в исходной строке. Слова должны быть отсортированы по убыванию их количества появления в тексте.

````python
import re

text = '''Дана строка, состоящая из слов, разделенных пробелами. Выведите все слова, встречающиеся в этой строке, на экран. 
Рядом выведите, сколько раз встречается слово в исходной строке. Слова должны быть отсортированы по убыванию их количества 
появления в тексте.'
'''.lower()

text = re.sub(r'[.,"\'-?:!;]', '', text)
text = text.split()
d = {}

for i in text: d[i] = 0
for i in text: d[i] +=1
    
def sort_by_value(x):
    return x[1]
    
sorted(d.items(), key=sort_by_value, reverse=True)
````

**Задание 2**

Имеется строка с названиями товаров вида «яблоки, груши, яблоки, киви, сливы, киви». Товары перечислены через запятую, товары могут повторяться. Выведите название товара, который встречается в этой строке чаще всего. Если таких товаров несколько, то названия всех товаров. 

````python
text = "яблоки, груши, яблоки, киви, сливы, киви".split(', ')
d = {}

for i in text: d[i] = 0
for i in text: d[i] +=1
    
max(d.values())
````

**Задание 3**

Число от 2 до 55, которое может содержать только цифры 2, 3, 4, 5 записано прописью, например, «тридцать три». Вычислите квадратный корень из этого числа.

````python
import math
calc = {
    "двадцадцать":20, 
    "два":2, 
    "тридцать":30, 
    "три":3, 
    "сорок":40, 
    "четыре":4, 
    "пятьдесят":50, 
    "пять":5 
}
count = 0
text = input("Введите число от 2 до 55 прописью: ").split()

for i in text:
    count = count+calc[i]
    
print("Квадратный корень из числа", " ".join(text), math.sqrt(count))
````

**Задание 6**

Определить сколько элементов заданного списка содержится в словаре. <br>
Пример: определить сколько элементов списка l1 содержится в словаре d1. <br>
```
l1 = list('abcd') d1 = {'c': 5, 'd': 6, 'e': 7, 'f': 8, 'g': 9, 'h': 10} 
```

````python
l1 = list('abcd') 
d1 = {'c': 5, 'd': 6, 'e': 7, 'f': 8, 'g': 9, 'h': 10}
count = 0

for i in range(len(l1)):
    for m in d1:
        if(l1[i]==m): count+=1
count
````
